```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
import math
# начальная точка отрезка
t_0 = 4
# конечная точка отрезка
T = 19
# количество интервалов, на которые будет разбит отрезок [t_0, T]; (Num+1) - количество узлов
Num = 3000

k_1=0    
k_2=10**5
k_3=10**(-16)
sigma=10**6

# начальные условия прямой задачи
u1_0 = 0
u2_0 = 1.3*(10**8)
u3_0 = 5.0*(10**11)
u4_0 = 8.0*(10**11)


# массив начальных условий прямой задачи
y_0 = np.zeros(4)
y_0[0] = u1_0
y_0[1] = u2_0
y_0[2] = u3_0
y_0[3] = u4_0


# вычисляет значение правой части прямой задачи в точке (t,y)
# y = (S, L, D, B, D0, B0)
def g(t, y):
    sec=(math.sin((1/16)*math.pi*(t-4)))**0.2
    k_1=10**(-5)*math.exp(7*sec)
    yP = np.zeros(4)
    yP[0] = k_1*y[2]-k_2*y[0]
    yP[1] = k_1*y[2]-k_3*y[1]*y[3]+sigma
    yP[2] = k_3*y[1]*y[3]-k_1*y[2]
    yP[3] = k_2*y[0]-k_3*y[1]*y[3]
    return yP

# метод Рунге-Кутты 4 порядка аппроксимации
# вычисляет значения искомой функции во всех узлах сетки
# n - количество интервалов, на которое разбивается отрезок [t0, t1]; (n+1) - количество узлов
# func(t,y) - правая часть системы; y'(t0)=y0
def rk4(func, t0, y0, t1, n):
    h = (t1 - t0) / n
    y = y0.copy()
    r = np.zeros((n + 1, 4))
    r[0] = y.copy()
    for i in range(1, n + 1):
        k1 = h * func(t0, y)
        k2 = h * func(t0 + 0.5 * h, y + 0.5 * k1)
        k3 = h * func(t0 + 0.5 * h, y + 0.5 * k2)
        k4 = h * func(t0 + h, y + k3)
        y = y + (1.0 / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
        r[i] = y.copy()
        t0 = t0 + h
    return r


x = np.arange(1, 12, 1)


grid = np.linspace(t_0,T, num=Num + 1, endpoint=True)
Y = rk4(g, t_0, y_0, T, Num)

fig, ax = plt.subplots()
#  Сплошная линия ('-' или 'solid',
#  установлен по умолчанию):
ax.plot(grid, Y[:,0],
        label='O',
        linewidth = 3,
        color = 'red')

# Пунктирная линия ('--' или 'dashed'):
ax.plot(grid, Y[:,1],
        label='NO',
        #linestyle = '--',
        linewidth = 3,
        color = 'yellow')

# Точка-тире ('-.' или 'dashdot'):
ax.plot(grid, Y[:,2], label='NO2',
        linewidth = 3,
        color = 'blue')

# Точка-точка (':' или 'dotted'):
ax.plot(grid, Y[:,3], label='O3',
        linewidth = 3,
        color = 'green')


fig.set_figwidth(12)
fig.set_figheight(6)
fig.set_facecolor('linen')
ax.set_facecolor('ivory')
plt.xlabel('время')
plt.ylabel('численность группы')
plt.title('Решение прямой задачи')

plt.legend()

plt.show()
```
