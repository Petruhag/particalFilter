```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt

# начальная точка отрезка
t_0 = 0
# конечная точка отрезка
T = 15
# количество интервалов, на которые будет разбит отрезок [t_0, T]; (Num+1) - количество узлов
Num = 3000
# количество измерений дополнительной информации
K = 5 # 5, 10, 15

# параметры модели
# q = (beta, k, p_D, gamma, alpha_D, beta_L, beta_B, beta_D, beta_L0, beta_B0, beta_D0)
f1 = 0.0
f2 = 0.0
f3 = 0.0

def f_s(t):
    return 2138.0

def f_L(t):
    return 1315.0

def f_D(t):
    return 26.0

def f_B(t):
    return 23.0

def f_D0(t):
    return 79.0

def f_B0(t):
    return 70.0

def beta_L0(x):
    return 0.443

def beta_D0(x):
    return 0.311

def beta_B0(x):
    return 5.72 * 10**(-2)

def phi_D(x):
    return 0.156

def phi_B(x):
    return 3.21

f = 10554.0
I_c = 0.3
beta = 4.5 * 10**(-5)
k = 0.1
mu = 1.74 * 10**(-2)
p_D = 5.0 * 10**(-2)
beta_L = 0.422
gamma = 1.43 * 10**(-3)
alpha_D = 2.25 * 10**(-6)
beta_D = 0.5
beta_B = 0.525
mu_D = 0.0533
mu_B = 0.5
mu_D0 = 5.33 * 10**(-2)
mu_B0 = 1.133

# начальные условия прямой задачи
S_0 = 277000.0
L_0 = 432000.0
D_0 = 900.0
B_0 = 118.0
D0_0 = 769.0
B0_0 = 944.0

# массив начальных условий прямой задачи
y_0 = np.zeros(6)
y_0[0] = S_0
y_0[1] = L_0
y_0[2] = D_0
y_0[3] = B_0
y_0[4] = D0_0
y_0[5] = B0_0


# вычисляет значение правой части прямой задачи в точке (t,y)
# y = (S, L, D, B, D0, B0)
def g(t, y):
    yP = np.zeros(6)
    yP[0] = f * (1.0 - I_c) - beta * y[0] * (y[3] + k * y[5]) - mu * y[0] + f_s(t)
    yP[1] = f * I_c + (1.0 - p_D) * beta * y[0] * (y[3] + k * y[5]) + beta_L * y[2] + beta_L0(f2) * y[4] \
            - y[1] * (gamma + alpha_D * (y[3] + k * y[5])) - mu * y[1] + f_L(t)
    yP[2] = p_D * beta * y[0] * (y[3] + k * y[5]) + y[1] * (gamma + alpha_D * (y[3] + k * y[5])) + beta_D * y[3] \
            - beta_B * y[2] - beta_L * y[2] - phi_D(f1) * y[2] - mu_D * y[2] + f_D(t)
    yP[3] = beta_B * y[2] - beta_D * y[3] - phi_B(f1) * y[3] - mu_B * y[3] + f_B(t)
    yP[4] = phi_D(f1) * y[2] + beta_D0(f3) * y[5] - beta_B0(f2) * y[4] - beta_L0(f2) * y[4] - mu_D0 * y[4] + f_D0(t)
    yP[5] = phi_B(f1) * y[3] + beta_B0(f2) * y[4] - beta_D0(f3) * y[5] - mu_B0 * y[5] + f_B0(t)
    return yP

# метод Рунге-Кутты 4 порядка аппроксимации
# вычисляет значения искомой функции во всех узлах сетки
# n - количество интервалов, на которое разбивается отрезок [t0, t1]; (n+1) - количество узлов
# func(t,y) - правая часть системы; y'(t0)=y0
def rk4(func, t0, y0, t1, n):
    h = (t1 - t0) / n
    y = y0.copy()
    r = np.zeros((n + 1, 6))
    r[0] = y.copy()
    for i in range(1, n + 1):
        k1 = h * func(t0, y)
        k2 = h * func(t0 + 0.5 * h, y + 0.5 * k1)
        k3 = h * func(t0 + 0.5 * h, y + 0.5 * k2)
        k4 = h * func(t0 + h, y + k3)
        y = y + (1.0 / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
        r[i] = y.copy()
        t0 = t0 + h
    return r

grid = np.linspace(t_0,T, num=Num + 1, endpoint=True)
Y = rk4(g, t_0, y_0, T, Num)


fig, ax = plt.subplots()

#  Сплошная линия ('-' или 'solid',
#  установлен по умолчанию):
ax.plot(grid, Y[:,0],
        label='s',
        linewidth = 3,
        color = 'crimson')

# Пунктирная линия ('--' или 'dashed'):
ax.plot(grid, Y[:,1],
        label='L',
        #linestyle = '--',
        linewidth = 3,
        color = 'darkmagenta')

# Точка-тире ('-.' или 'dashdot'):
ax.plot(grid, Y[:,2], label='D',
        linewidth = 3,
        color = 'indigo')

# Точка-точка (':' или 'dotted'):
ax.plot(grid, Y[:,3], label='B',
        linewidth = 3,
        color = 'darkblue')

ax.plot(grid, Y[:,4], label='D0',
        #linestyle = ':',
        linewidth = 3,
        color = 'darkgreen')

ax.plot(grid, Y[:,5], label='B0',
        #linestyle = ':',
        linewidth = 3,
        color = 'darkred')

fig.set_figwidth(12)
fig.set_figheight(6)
fig.set_facecolor('linen')
ax.set_facecolor('ivory')
plt.xlabel('время')
plt.ylabel('численность группы')
plt.title('Решение прямой задачи')

plt.legend()

plt.show()
```
