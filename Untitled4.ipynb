```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt

#  c         
def findmax(C):
    sum_max = 0
    cl_max = 0
    for j in range(len(C[0])):
        sum_column = 0
        for i in range(len(C)):
            sum_column = sum_column + (C[i][j]) ** 2
        if sum_column >= sum_max:
            sum_max = sum_column
            cl_max = j
    return cl_max

#    E    c_max
#      S
#        I
#      U
def add_delete(Sr, E, I, U, c_max):
    d = Sr[1:len(Sr), c_max].copy()
    d1 = d[:, np.newaxis]
    E = np.hstack((E, d1))
    p = int(Sr[0][c_max])
    Sr = np.delete(Sr, c_max, axis=1)
    I.append(p)
    U.remove(p)
    return [Sr, E, I, U]

#   
t_0 = 0
#   
T = 15
#  ,      [t_0, T]; (Num+1) -  
Num = 3000
#    
K = 15 # 5, 10, 15

#  
# q = (beta, k, p_D, gamma, alpha_D, beta_L, beta_B, beta_D, beta_L0, beta_B0, beta_D0)
f1 = 0.0
f2 = 0.0
f3 = 0.0

def f_s(t):
    return 2138.0

def f_L(t):
    return 1315.0

def f_D(t):
    return 26.0

def f_B(t):
    return 23.0

def f_D0(t):
    return 79.0

def f_B0(t):
    return 70.0

def beta_L0(x):
    return 0.443

def beta_D0(x):
    return 0.311

def beta_B0(x):
    return 5.72 * 10**(-2)

def phi_D(x):
    return 0.156

def phi_B(x):
    return 3.21

f = 10554.0
I_c = 0.3
beta = 4.5 * 10**(-5)
k = 0.1
mu = 1.74 * 10**(-2)
p_D = 5.0 * 10**(-2)
beta_L = 0.422
gamma = 1.43 * 10**(-3)
alpha_D = 2.25 * 10**(-6)
beta_D = 0.5
beta_B = 0.525
mu_D = 0.0533
mu_B = 0.5
mu_D0 = 5.33 * 10**(-2)
mu_B0 = 1.133

#    
S_0 = 277000.0
L_0 = 432000.0
D_0 = 900.0
B_0 = 118.0
D0_0 = 769.0
B0_0 = 944.0

#     
y_0 = np.zeros(6)
y_0[0] = S_0
y_0[1] = L_0
y_0[2] = D_0
y_0[3] = B_0
y_0[4] = D0_0
y_0[5] = B0_0

#    T_K = (t_1, ..., t_K),    
T_K = np.arange(t_0 + 1, K + 1, 1)

#         (t,y)
# y = (S, L, D, B, D0, B0)
def g(t, y):
    yP = np.zeros(6)
    yP[0] = f * (1.0 - I_c) - beta * y[0] * (y[3] + k * y[5]) - mu * y[0] + f_s(t)
    yP[1] = f * I_c + (1.0 - p_D) * beta * y[0] * (y[3] + k * y[5]) + beta_L * y[2] + beta_L0(f2) * y[4] \
            - y[1] * (gamma + alpha_D * (y[3] + k * y[5])) - mu * y[1] + f_L(t)
    yP[2] = p_D * beta * y[0] * (y[3] + k * y[5]) + y[1] * (gamma + alpha_D * (y[3] + k * y[5])) + beta_D * y[3] \
            - beta_B * y[2] - beta_L * y[2] - phi_D(f1) * y[2] - mu_D * y[2] + f_D(t)
    yP[3] = beta_B * y[2] - beta_D * y[3] - phi_B(f1) * y[3] - mu_B * y[3] + f_B(t)
    yP[4] = phi_D(f1) * y[2] + beta_D0(f3) * y[5] - beta_B0(f2) * y[4] - beta_L0(f2) * y[4] - mu_D0 * y[4] + f_D0(t)
    yP[5] = phi_B(f1) * y[3] + beta_B0(f2) * y[4] - beta_D0(f3) * y[5] - mu_B0 * y[5] + f_B0(t)
    return yP

#  - 4  
#        
# n -  ,     [t0, t1]; (n+1) -  
# func(t,y) -   ; y'(t0)=y0
def rk4(func, t0, y0, t1, n):
    h = (t1 - t0) / n
    y = y0.copy()
    r = np.zeros((n + 1, 6))
    r[0] = y.copy()
    for i in range(1, n + 1):
        k1 = h * func(t0, y)
        k2 = h * func(t0 + 0.5 * h, y + 0.5 * k1)
        k3 = h * func(t0 + 0.5 * h, y + 0.5 * k2)
        k4 = h * func(t0 + h, y + k3)
        y = y + (1.0 / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
        r[i] = y.copy()
        t0 = t0 + h
    return r

#   dg/dy    y
def dgdy(y):
    G = np.zeros((6,6))
    G[0][0] = - (beta * (y[3] + k * y[5]) + mu)
    G[0][3] = - beta * y[0]
    G[0][5] = - beta * y[0] * k
    G[1][0] = (1.0 - p_D) * beta * (y[3] + k * y[5])
    G[1][1] = - (gamma + alpha_D * (y[3] + k * y[5]) + mu)
    G[1][2] = beta_L
    G[1][3] = (1.0 - p_D) * beta * y[0] - y[1] * alpha_D
    G[1][4] = beta_L0(f2)
    G[1][5] = ((1.0 - p_D) * beta * y[0] - y[1] * alpha_D) * k
    G[2][0] = p_D * beta * (y[3] + k * y[5])
    G[2][1] = gamma + alpha_D * (y[3] + k * y[5])
    G[2][2] = - (beta_B + beta_L + phi_D(f1) + mu_D)
    G[2][3] = p_D * beta * y[0] + y[1] * alpha_D + beta_D
    G[2][5] = (p_D * beta * y[0] + y[1] * alpha_D) * k
    G[3][2] = beta_B
    G[3][3] = - (beta_D + phi_B(f1) + mu_B)
    G[4][2] = phi_D(f1)
    G[4][4] = - (beta_B0(f2) + beta_L0(f2) + mu_D0)
    G[4][5] = beta_D0(f3)
    G[5][3] = phi_B(f1)
    G[5][4] = beta_B0(f2)
    G[5][5] = - (beta_D0(f3) + mu_B0)
    return G

#   (dg/dq)^T    y
def dgdqT(y):
    G = np.zeros((6, 11))
    G[0][0] = - y[0] * (y[3] + k * y[5])
    G[0][1] = - beta * y[0] * y[5]
    G[1][0] = (1.0 - p_D) * y[0] * (y[3] + k * y[5])
    G[1][1] = ((1.0 - p_D) * beta * y[0] - y[1] * alpha_D) * y[5]
    G[1][2] = - beta * y[0] * (y[3] + k * y[5])
    G[1][3] = - y[1]
    G[1][4] = - y[1] * (y[3] + k * y[5])
    G[1][5] = y[2]
    G[1][8] = y[4]
    G[2][0] = p_D * y[0] * (y[3] + k * y[5])
    G[2][1] = (p_D * beta * y[0] + y[1] * alpha_D) * y[5]
    G[2][2] = beta * y[0] * (y[3] + k * y[5])
    G[2][3] = y[1]
    G[2][4] = y[1] * (y[3] + k * y[5])
    G[2][5] = - y[2]
    G[2][6] = - y[2]
    G[2][7] = y[3]
    G[3][6] = y[2]
    G[3][7] = - y[3]
    G[4][8] = - y[4]
    G[4][9] = - y[4]
    G[4][10] = y[5]
    G[5][9] = y[4]
    G[5][10] = - y[5]
    return G.T

# c  s_q_j   TK = (t_1, ..., t_K)  - 4 
#             [t0, t1]
def rk4Sq(func, t0, y0, t1, n, TK, j):
    h = (t1 - t0) / n
    K0 = TK.size
    r = np.zeros((K0, 6))
    Y = rk4(func, t0, y0, t1, 2*n)
    s = np.zeros(6)
    m = 0
    for i in range(1, n+1):
        p = 2 * (i - 1)
        A0 = dgdy(Y[p])
        b0 = dgdqT(Y[p])[j]
        A12 = dgdy(Y[p + 1])
        b12 = dgdqT(Y[p + 1])[j]
        A1 = dgdy(Y[p + 2])
        b1 = dgdqT(Y[p + 2])[j]
        k1 = h * (np.matmul(A0, s) + b0)
        k2 = h * (np.matmul(A12, s + 0.5 * k1) + b12)
        k3 = h * (np.matmul(A12, s + 0.5 * k2) + b12)
        k4 = h * (np.matmul(A1, s + k3) + b1)
        s = s + (1.0 / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
        t0 = t0 + h
        if abs(t0 - TK[m]) < 10**(-3):
            r[m] = s.copy()
            if m < K0-1:
                m += 1
    return r

dhdy = np.zeros((3, 6))
dhdy[0][0] = 1.0
dhdy[1][4] = 1.0
dhdy[2][5] = 1.0

dhdq = np.zeros((3, 11))

#    S
def Sen(func, t0, y0, t1, n, TK):
    K0 = TK.size
    S = np.zeros((3*K0, 11))
    for j in range(11):
        Sq = rk4Sq(func, t0, y0, t1, n, TK, j)
        for i in range(3):
            for k in range(K0):
                S[i + 3*k][j] = np.dot(dhdy[i], Sq[k]) + dhdq[i][j]
    return S


S = Sen(g, t_0, y_0, T, Num, T_K)
print('S')
print(S)

#    S
m = len(S)
#    S   
L = len(S[0])

I = []  # 
U = []  # 

for i in range(L):
    U.append(i)

#  c     
col_max = findmax(S)
print('col_max')
print(col_max)

#       
S_max = np.zeros(L)
S_max[0] = norm(S[:, col_max], ord=2)

SR = np.zeros((m + 1, L))
for i in range(L):
    SR[0][i] = i
SR[1:m + 1] = S.copy()
print('SR')
print(SR)

E = np.zeros((m, 0))

#                 S
temp = add_delete(SR, E, I, U, col_max)
SR = temp[0]
E = temp[1]
I = temp[2]
U = temp[3]

S_proj = np.zeros((m, 1))
S_perpend = np.zeros((m, L))

while True:
    if len(U) != 0:
        n = len(SR[0])
        S_perpend = np.zeros((m, n))
        for h in range(n):
            for k in range(L - n):
                S_proj = S_proj + (np.dot(SR[1:m + 1, h], E[:, k]) / np.dot(E[:, k], E[:, k])) * E[:, k]
            S_perpend[:, h] = S[:, h] - S_proj[:, 0]
        #     c     
        col_max = findmax(S_perpend)
        S_max[L - n] = norm(S_perpend[:, col_max], ord=2)
        temp = add_delete(SR, E, I, U, col_max)
        SR = temp[0]
        E = temp[1]
        I = temp[2]
        U = temp[3]
    else:
        break

print('S_max')
print(S_max)

x = np.arange(1, 12, 1)

chars = ['beta', 'k', 'p_D', 'gamma', 'alpha_D', 'beta_L', 'beta_B', 'beta_D', 'beta_L0', 'beta_B0', 'beta_D0']
chars1 = []
for i in I:
    chars1.append(chars[i])

print(I)
print(chars1)

fig, ax = plt.subplots()

ax.bar(x, np.log10(S_max))
ax.set_xticks(x)
ax.set_xticklabels(chars1)

ax.set_facecolor('seashell')
fig.set_facecolor('floralwhite')
fig.set_figwidth(12)    #   Figure
fig.set_figheight(6)    #   Figure
plt.title('   ( )')


#plt.hist(np.log(S_max), density=True, alpha=1, color="k")
#plt.ylabel('norm_perpend')
#plt.xlabel('parameters')
plt.show()

grid = np.linspace(t_0,T, num=Num + 1, endpoint=True)
Y = rk4(g, t_0, y_0, T, Num)

fig, ax = plt.subplots()
#    ('-'  'solid',
#    ):
ax.plot(grid, Y[:,0],
        label='s',
        linewidth = 3,
        color = 'crimson')

#   ('--'  'dashed'):
ax.plot(grid, Y[:,1],
        label='L',
        #linestyle = '--',
        linewidth = 3,
        color = 'darkmagenta')

# - ('-.'  'dashdot'):
ax.plot(grid, Y[:,2], label='D',
        linewidth = 3,
        color = 'indigo')

# - (':'  'dotted'):
ax.plot(grid, Y[:,3], label='B',
        linewidth = 3,
        color = 'darkblue')

ax.plot(grid, Y[:,4], label='D0',
        #linestyle = ':',
        linewidth = 3,
        color = 'darkgreen')

ax.plot(grid, Y[:,5], label='B0',
        #linestyle = ':',
        linewidth = 3,
        color = 'darkred')

fig.set_figwidth(12)
fig.set_figheight(6)
fig.set_facecolor('linen')
ax.set_facecolor('ivory')
plt.xlabel('')
plt.ylabel(' ')
plt.title('  ')

plt.legend()

plt.show()
```
